```python
def latest_by_country(data):
    return data.groupby("Country/Region").sum().iloc[:,-1]

information = [latest_by_country(confirmed),latest_by_country(mortes),latest_by_country(recuperados)]
information
```

```python
combined = pd.concat(information, axis=1)
combined.columns  = ["Confirmados","Mortos","Recuperados"]
combined
```

```python
sum_up = combined.sum()
letalidade = sum_up["Mortos"] / sum_up["Confirmados"]
print((letalidade*100))
```

```python
letalidade = combined["Mortos"] / combined["Confirmados"]*100
combined["Letalidade_1"] = letalidade
combined.head(95)
```

```python
# Tabela de Casos recuperados não está organizada como as tabelas anteriores, não exibindo a soma e sim um valor absoluto por dia
recuperados_por_pais = recuperados.groupby("Country/Region").sum()
recuperados_por_pais.drop(columns="Lat", inplace=True)
recuperados_por_pais.drop(columns="Long", inplace=True)
recuperados_por_pais.sum(axis=1)
```

```python
def latest_by_country_at(data, date):
    return data.groupby("Country/Region").sum()[date]

informations = [latest_by_country_at(confirmed, '2/20/20'),latest_by_country_at(mortes, '2/20/20'),latest_by_country_at(recuperados, '2/20/20')]
combined_2_20_20 = pd.concat(informations, axis=1)
combined_2_20_20.columns = ["Confirmed", "Deaths", "Recovered"]

sum_up = combined_2_20_20.loc["China"]
letality_rate_1 = sum_up["Deaths"] / sum_up["Confirmed"]
print(letality_rate_1 * 100)
```

```python

#Exemplo de calculo com taxa de letalidade móvel ( Mortos_dia_X / Confirmados_Dia_X[-T]) onde T é o tempo médio de infeccao

informations = [latest_by_country_at(confirmed, '2/8/20'),latest_by_country_at(mortes, '2/20/20'),latest_by_country_at(recuperados, '2/20/20')]
combined12 = pd.concat(informations, axis=1)
combined12.columns = ["Confirmed", "Deaths", "Recovered"]

sum_up = combined12.loc["China"]
letality_rate_1 = sum_up["Deaths"] / sum_up["Confirmed"]
display(sum_up)
print(letality_rate_1 * 100)
```
