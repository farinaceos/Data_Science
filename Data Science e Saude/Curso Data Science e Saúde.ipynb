```python
def latest_by_country_at(data, date):
    return data.groupby("Country/Region").sum()[date]

informations = [latest_by_country_at(confirmed, '2/20/20'),latest_by_country_at(mortes, '2/20/20'),latest_by_country_at(recuperados, '2/20/20')]
combined_2_20_20 = pd.concat(informations, axis=1)
combined_2_20_20.columns = ["Confirmed", "Deaths", "Recovered"]

sum_up = combined_2_20_20.loc["China"]
letality_rate_1 = sum_up["Deaths"] / sum_up["Confirmed"]
print(letality_rate_1 * 100)
```

```python
un_population = pd.read_csv("dados/Pop_UN.csv")
un_population["PopTotal"] = un_population["PopTotal"] * 1000 #Dados estão na escala dos milhares, mantendo a escala igual para o Dataset anterior
Countries_Population = un_population.query("Time == 2019")[["Location","PopTotal"]].set_index("Location")
```

```python
combined.join(Countries_Population).loc["Korea, South"]
```

```python
un_population.query("Location.str.contains('Iran') & Time == 2019")
```

```python
#Renomeando Locations para conseguir unir os dois Datasets
def rename_location(location):
    if location == 'Republic of Korea':
        return 'Korea, South'
    if location == 'United States of America':
        return 'US'
    if location == 'Russian Federation':
        return 'Russia'
    if location == 'Türkiye':
        return 'Turkey'
    if location == 'Viet Nam':
        return 'Vietnam'
    if location == 'China, Taiwan Province of China':
        return 'Taiwan*'
    if location == 'Iran (Islamic Republic of)':
        return 'Iran'
    
    return location

Countries_Population['Location_for_who'] = Countries_Population.index.map(rename_location)
```

```python
Countries_Population = Countries_Population.set_index("Location_for_who")
Countries_Population.head()
```

```python
hubei_population = 59020000
confirmed_cases = confirmed.set_index("Province/State").loc["Hubei"][last_date]
death_cases = mortes.set_index("Province/State").loc["Hubei"][last_date]
recovered_cases = recuperados.set_index("Province/State").loc["Hubei"][last_date]

let_rate_1 = death_cases / confirmed_cases * 100
let_rate_2 = death_cases / (death_cases + recovered_cases) * 100

print(f"Letality Rate 1: {let_rate_1.round(2)}%")
print(f"Letality Rate 1: {let_rate_2.round(2)}%")
```

```python
incidence_ratio = confirmed_cases / hubei_population * 100000
mortality_rate = death_cases / hubei_population * 100000
print(f"Coeficiente de incidência em Hubei {incidence_ratio.round(2)}")
print(f"Coeficiente de mortalidade em Hubei {mortality_rate.round(2)}")
```

```python
grouped_per_day = pd.DataFrame([confirmed.sum()[3:-1],recuperados.sum()[1:], mortes.sum()[3:]],
            index=["Confirmados","Recuperados", "Mortos"]).T
grouped_per_day.fillna(method='ffill')
```

```python
grouped_per_day["Letality_Rate_1"] = grouped_per_day["Mortos"] / grouped_per_day["Confirmados"] * 100
grouped_per_day["Letality_Rate_2"] = grouped_per_day["Mortos"] /( grouped_per_day["Mortos"] + grouped_per_day["Recuperados"]) * 100
grouped_per_day.head()
```

```python
plt.figure(figsize=(10,5))
ax = grouped_per_day["Letality_Rate_1"].plot()
ax = grouped_per_day["Letality_Rate_2"].plot()
plt.title("Taxa de Letalidade 'Pessimista' e 'Otimista' no mundo todo")
plt.legend(['Mortes / Casos Confirmados', 'Mortes / (Mortes + Recuperados)'])
ax.set_xlabel("Fonte: Curso www.alura.com.br e dados Johns Hopkins CSSE https://github.com/CSSEGISandData/COVID-19")
ax.set_ylabel("% Letalidade")
plt.grid(linestyle = '-')
plt.show()
```
